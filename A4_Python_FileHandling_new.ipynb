```python
#Assignment 4: File Handling (Read, Write, Append, Modes)

#Task 1: Write Sales Records to a File 

sales = [1200, 450, 980, 1500, 3000]
#write
with open('sales_data.txt','w') as file:
    for sale in sales:
        file.write(str(sale)+'\n')

print ('Sales data written to file succesfully')

#read
with open('sales_data.txt','r') as file:
    for sales in file:
        print(sales)
```

```python
#Comma-Separated Format
sales = [1200, 450, 980, 1500, 3000]
#write
with open('sales_comma_data.txt','w') as file:
    str_sales=list(map(str,sales))
    #comma separate
    file.write(','.join(str_sales))
print('Data written succesfully')

with open('sales_comma_data.txt','r') as file:
    print(file.read())
```

```python
#Task 2: Read File in Different Ways 

#read entire file
with open('sales_data.txt','r') as file:
    print(file.read())

#read first line
with open('sales_data.txt','r') as file:
    print('reading first line :',file.readline().strip())

#read all lines using readlines
with open('sales_data.txt','r') as file:
    read_line=[int(line.strip()) for line in file.readlines()]
for i in read_line:
    print(i,end=',')
    
    
```

```python
#Task 3: Append New Sales 

#append new sales to the same file.
new_sales=[5000,2500,1700]
with open('sales_data.txt','a') as file:
    for new in new_sales:
        file.write(str(new)+'\n')
print ('new data written succesfully')

# reading the data (contains new data)
with open('sales_data.txt','r')as file:
    print(file.read())

# print total number of lines
with open('sales_data.txt','r')as file:
    print(len(file.readlines()))
```

```python
#Task 4: Generate Summary Report from File 
with open('sales_data.txt','r') as file:
    lis=list(int(line.strip()) for line in file.readlines())
print('=============Summary Report=============')
print ('sales:',lis)
print('total sales:',sum(lis))
print('Highest sale:',max(lis))
print('Lowest sale:',min(lis))
print('Average sales:', sum(lis)/len(lis))
```

```python
#Task 5: Create Product Info File (User Input) 
 
with open('product.txt','w') as file:
    print('Enter 3 product details')
    for i in range(1,4):
        name=input(f"Enter name of Product {i}: ").strip().capitalize()
        price=input(f"Enter price of Product {i}: ").strip()

        file.write(f'{name} | {price} \n')
    print('Data written successful')
```

```python
with open('product.txt','r') as file:
    print('=============Product Details================')
    print(f"{'Product Name':<20} | {'Product Price':>10}")
    print('============================================')
    for line in file:
        parts=line.split('|')
        productName=parts[0].strip()
        productPrice=parts[1].strip()

        print(f'{productName:<20} | {productPrice}')
```

```python
#Task 6: Read File Safely (Error Handling Inside File Handling Only)
import os
filename=input('Enter the file name you want to open: ').strip()

if os.path.exists(filename):
    print(f"--------Reading content of {filename}------------")

    with open(filename,'r') as file:
        print(file.read())
        
else:
    print("File not found")
```

```python
#Task 7: Mini Project - Export Discounted Prices 
prices = { 
"Mouse": 500, 
"Keyboard": 800, 
"Monitor": 7000, 
"Pendrive": 400, 
"Camera": 5000}

dis_per=int(input('Enter discount percentage : '))
print ("Discount percentage is :", dis_per)
discount_sum=0
with open('discount_report.txt','w') as file:
    file.write(f"{'Product':<10} | {'Original Price':>15} | {'Discounted Price':>20}\n")
    file.write("-" * 52 + "\n")
    for product,price in prices.items():
        discount=price-(price * (dis_per / 100))
        discount_sum += discount
        file.write(f"{product:<10} | {price:>15} | {discount:>20}\n")

    avg_discount=discount_sum/len(prices)
    file.write("-" * 52 + "\n")
    file.write("============= SUMMARY REPORT =============\n")
    file.write(f"total items: {len(prices)}\n")
    file.write(f'Average Discounted price: {avg_discount}\n')
        
print("Data written succesfully")

with open('discount_report.txt','r') as file:
    print(file.read())
```
